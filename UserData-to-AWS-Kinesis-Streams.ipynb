```python
import requests
import boto3
import uuid
import time
import random
import json
print("done")
```

```python
client = boto3.client('kinesis', region_name='us-east-1')
partition_key = str(uuid.uuid4())
random_user_url = 'https://randomuser.me/api/?exc=login'
```

```python
while True:
        r = requests.get(random_user_url)
        data = json.dumps(r.json())
        client.put_record(
                StreamName='user_data_stream1',
                Data=data,
                PartitionKey=partition_key)
        print(data)
        time.sleep(random.uniform(0, 1))
```
